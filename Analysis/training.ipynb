```python
import numpy as np
import matplotlib.pyplot as plt
import os, sys

cmd = os.path.dirname(sys.path[0])
print(cmd)
sys.path.append(os.path.join(cmd,"3DTomoGAN"))
sys.path.append(os.path.join(cmd,"Results"))


%reload_ext autoreload
%autoreload 2
```

```python
folder = r"F:\Repos\4D_CT\3DTomoGAN\single_experimental_long-itrOut"

train_loss = "train_loss.npy"
val_loss = "train_loss.npy"

mse_part = "mse_list.npy"
adv_part = "adv_list.npy"

x = "x_axis.npy"

```

```python
loading_npy = lambda f, n: np.load(os.path.join(f, n))
train = loading_npy(folder, train_loss)
val = loading_npy(folder, val_loss)

mse = loading_npy(folder, mse_part)
adv = loading_npy(folder, adv_part)

x = loading_npy(folder, x)
```

```python
fig, ax = plt.subplots(1,2)

ax[0].plot(train, label = "train", alpha=0.5)
ax[0].plot(x,  val, label="val", alpha=0.5)

ax[1].plot(x, mse, label="mse", alpha=0.5)
ax[1].plot(x, adv, label="adv", alpha=0.5)

ax[0].legend()
ax[1].legend()

ax[0].set_ylim(0,50)
ax[0].set_xlim(4000,6000)
ax[1].set_ylim(0,50)
ax[1].set_xlim(4000,6000)
plt.show()

print(len(train))
print(len(val))
```
