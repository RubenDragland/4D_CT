```python
def recursively_dict_value(keys, value):
    # print("len keys", len(keys))
    if len(keys) == 1:
        # dict[keys[0]] = value
        # return dict
        return {keys[0]: value}
    else:
        return {keys[0]: recursively_dict_value( keys[1:], value)}
    
```
